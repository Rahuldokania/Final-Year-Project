```python
import numpy as np
import pandas as pd 
import warnings 
warnings.filterwarnings('ignore')
```

```python
data=pd.read_excel("D://downloads//Participants_Data_Used_Cars//Data_Train.xlsx")
```

```python
data.head(6)
```

```python
data.info()
```

```python
data['Name']=data['Name'].str.split(' ').str[0].str.lower()#+data['Name'].str.split(' ').str[1]+data['Name'].str.split(' ').str[2]
data['Location']=data['Location'].str.lower()
data['Fuel_Type']=data['Fuel_Type'].str.lower()
data['Transmission']=data['Transmission'].str.lower()
data['Owner_Type']=data['Owner_Type'].str.lower()
```

```python
print(data['Name'].unique())
print(data['Location'].unique())
print(data['Fuel_Type'].unique())
print(data['Transmission'].unique())
print(data['Owner_Type'].unique())
```

```python
data['Mileage']=data['Mileage'].str.replace(' km/kg','').astype(float,errors='ignore')
data['Mileage']=data['Mileage'].str.replace(' kmpl','').astype(float)
data['Engine']=data['Engine'].str.replace(' CC','').astype(float,errors='ignore')
data['Power']=data['Power'].str.replace(' bhp','').astype(float,errors='ignore')
data['Power']=pd.to_numeric(data['Power'],errors='coerce')
```

```python
data['Fuel_Type']=data.Fuel_Type.astype("category").cat.codes

#CNG =0  Diesel=1 Petrol=4 LPG=3 Electric=2
data['Transmission']=data.Transmission.astype("category").cat.codes
#Automatic=0 Manual=1
```

```python
from sklearn.preprocessing import LabelEncoder 
labelencoder= LabelEncoder() #initializing an object of class LabelEncoder
data['Name'] = labelencoder.fit_transform(data['Name']) 
data['Location'] = labelencoder.fit_transform(data['Location'])
data['Owner_Type'] = labelencoder.fit_transform(data['Owner_Type'])
data.head()
```

```python
total = data.isnull().sum().sort_values(ascending=False)
percent_1 = data.isnull().sum()/data.isnull().count()*100
percent_2 = (round(percent_1, 1)).sort_values(ascending=False)
missing_data = pd.concat([total, percent_2], axis=1, keys=['Total', '%'])
missing_data.head(6)
```

```python
data.info()
```

```python
data['Mileage']=data['Mileage'].fillna(data['Mileage'].mean())
data['Engine']=data['Engine'].fillna(data['Engine'].mean())
data['Power']=data['Power'].fillna(data['Power'].mean())
data['Seats']=data['Seats'].fillna(data['Seats'].mean())
data.head()
```

```python
X=data.iloc[:,[2,3,4,5,7,8,9]]
y=data.iloc[:,12]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0) 
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression
regressor1 = LinearRegression(normalize=True)
regressor1.fit(X_train, y_train)
y_pred1 = regressor1.predict(X_test) 
y_pred1
acc_lm = round((regressor1.score(X_test, y_test) * 100), 2)
```

Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
regressor2 = RandomForestRegressor(n_estimators=35, random_state=42)  
regressor2.fit(X_train, y_train)  
y_pred2 = regressor2.predict(X_test) 
y_pred2
acc_rf = round((regressor2.score(X_test, y_test) * 100), 2)
```

Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
regressor3 = DecisionTreeRegressor(random_state=42)
regressor3.fit(X_train, y_train)  
y_pred3 = regressor3.predict(X_test) 
y_pred3
acc_dt = round((regressor3.score(X_test, y_test) * 100), 2)
```

```python
results = pd.DataFrame({
    'Model': ['Linear Regressor', 'Random Forest Regressor', 'Decision Tree Regressor',],
    'Score': [acc_lm, acc_rf, acc_dt,]})
result_df = results.sort_values(by='Score', ascending=False)
result_df = result_df.set_index('Score')
result_df.head()
```

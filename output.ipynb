```python
def output():
    import joblib
    import numpy as np
    import pandas as pd
    from termcolor import colored,cprint
    import warnings 
    warnings.filterwarnings('ignore')
    load_model= joblib.load("Final.pkl")
    
    Brand=input('Enter the Brand:')

    Model=input('Enter the Model of the car: ')

    Location=input('Enter the city name:')

    Seats=int(input('Enter the no.of Seats:'))

    Year=int(input('Enter the year range(1998,2019):'))
    if (Year>2019 or Year<1998):
        cprint('Error in Year','grey','on_red',attrs=['bold'])
        raise ValueError

    Kilometers_Driven=float(input('Enter the kilometers driven :'))

    Fuel_Type=int(input('Enter the fuel type for (CNG =0  Diesel=1 Electric=2 LPG=3 Petrol=4) :'))
    #CNG =0  Diesel=1 Electric=2 LPG=3 Petrol=4 
    if (Fuel_Type > 4 or Fuel_Type<0):
        cprint('Error in Fuel_Type','grey','on_red',attrs=['bold'])
        raise ValueError

    Transmission=int(input('Enter the transmission type for (Automatic=0 Manual=1) :'))
    #Automatic=0 Manual=1
    if (Transmission > 1 or Transmission < 0):
        cprint('Error in Transmission','grey','on_red',attrs=['bold'])
        raise ValueError

    Mileage=float(input('Enter the Mileage in kmpl range(0,45):'))
    #kmpl
    if (Mileage > 45 or Mileage < 0):
        cprint('Error in Mileage','grey','on_red',attrs=['bold'])
        raise ValueError

    Engine=float(input('Enter the Engine in CC range(70,6000):'))
    #CC
    if (Engine > 6000 or Engine < 70):
        cprint('Error in Engine','grey','on_red',attrs=['bold'])
        raise ValueError

    Power=float(input('Enter the Power in bph range(34,560):'))
    #bhp
    if (Power > 560 or Power < 34):
        cprint('Error in Power','grey','on_red',attrs=['bold'])
        raise ValueError
        
        
    test=np.array([Year,Kilometers_Driven,Fuel_Type,Transmission,Mileage,Engine,Power]).reshape(1,-1)
    
    
    Price=load_model.predict(test)
    df=pd.DataFrame(Price)
    df.columns = ['Price in Lakhs']
    cprint('The expected Price of the car in Lakhs is: ',attrs=['bold'])
    cprint(round(Price[0],3),'grey','on_yellow',attrs=['bold'])
   
    
```
